# K-Means Clustering

## Importing the libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:,3:].values
```

## Using the elbow method to find the optimal number of clusters

```
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters = i,init ='k-means++',random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_) # inertia_ stores wcss of the model
plt.plot(range(1,11),wcss)
plt.scatter(range(1,11),wcss,color = 'red')
plt.show()
```

## Training the K-Means model on the dataset

```
km = KMeans(n_clusters = 5, init = 'k-means++', random_state=42)
y_pred = km.fit_predict(X)
print(y_pred)
```

## Visualising the clusters

```
colors =['red','green','blue','yellow','black']
plt.scatter(X[y_pred==0,0], X[y_pred==0,1],s = 100, c=colors[0], label = 'Cluster1')
plt.scatter(X[y_pred==1,0], X[y_pred==1,1],s = 100, c=colors[1], label = 'Cluster2')
plt.scatter(X[y_pred==2,0], X[y_pred==2,1],s = 100, c=colors[2], label = 'Cluster3')
plt.scatter(X[y_pred==3,0], X[y_pred==3,1],s = 100, c=colors[3], label = 'Cluster4')
plt.scatter(X[y_pred==4,0], X[y_pred==4,1],s = 100, c=colors[4], label = 'Cluster5')
#plotting centroids
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s = 300, color = 'violet', label = 'centroid')
plt.legend()
plt.show()
```
